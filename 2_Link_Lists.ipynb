```python
random.seed(42)

# Create 2 different lists of 500 bigrams, all containing the word 'fintech'
bigrams1 = []
bigrams2 = []

for i in range(500):
    search_str_1 = f'fintech {random.choice(glossary_lemmed)}'
    bigrams1.append(search_str_1)
    
    search_str_2 = f'{random.choice(glossary_lemmed)} fintech'
    bigrams2.append(search_str_2)
    

# Create 6 lists of 400 trigrams, all containing the word 'fintech'
trigrams1 = []
trigrams2 = []
trigrams3 = []
trigrams4 = []
trigrams5 = []
trigrams6 = []

for i in range(400):
    search_str_1 = f'fintech {random.choice(glossary_lemmed)} {random.choice(glossary_lemmed)}'
    trigrams1.append(search_str_1)
    
    search_str_2 = f'fintech jobs {random.choice(glossary_lemmed)}'
    trigrams2.append(search_str_2)
    
    search_str_3 = f'{random.choice(glossary_lemmed)} fintech recruitment'
    trigrams3.append(search_str_3)
    
    search_str_4 = f'{random.choice(glossary_lemmed)} {random.choice(glossary_lemmed)} fintech'
    trigrams4.append(search_str_4)
    
    search_str_5 = f'fintech dublin {random.choice(glossary_lemmed)}'
    trigrams5.append(search_str_5)
    
    search_str_6 = f'fintech {random.choice(glossary_lemmed)} london'
    trigrams6.append(search_str_6)
```

```python
######## New glossary files ########

to_pickle = {'bigrams1': bigrams1, 'bigrams2': bigrams2,
             'trigrams1': trigrams1, 'trigrams2': trigrams2,
             'trigrams3': trigrams3, 'trigrams4': trigrams4,
             'trigrams5': trigrams5, 'trigrams6': trigrams6}

# for lst_name, lst in to_pickle.items():
#     pickle.dump(lst, open(f'pickles/{lst_name}_glossary.pkl', 'wb'))
```

```python
def get_links(query):
    links = []
    for i in search(query, tld='com', num=10, stop=10, pause=2):
        links.append(i)
        print(i)
    return links
```

```python
# %%time

# final_links = [get_links(item) for item in glossary_lemmed]

# np.savetxt('links_list_lemmed.txt', final_links, delimiter="\n", fmt="%s")
```

```python
for lst_name, lst in to_pickle.items():
    # loop over all of the lists in the to_pickle dictionary and execute the 
    # get_links() function in each element of the current list
    start = time.time()
    final_links = [get_links(item) for item in lst]
    end = time.time()
    
    np.savetxt(f'links_list_{lst_name}_lemmed.txt', final_links, delimiter="\n", fmt="%s")
    
    print(f'''Finished getting the links for the search terms in {lst_name}.
    Took {(end-start)/60} minutes to complete.
    Starting the next search terms list...''')
```
